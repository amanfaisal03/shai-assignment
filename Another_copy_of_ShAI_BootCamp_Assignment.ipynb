```python
#The job title is a categorical feature that cannot be used in my programming. To address this, convert it to numerical values using label encoding or one-hot encoding.
lable_encod=preprocessing.LabelEncoder()
df['JobTitle']=lable_encod.fit_transform(df['JobTitle'])
df['Agency']=lable_encod.fit_transform(df['Agency'])
df.head()
```

```python
df.describe()
```

```python
df.drop("Notes", axis='columns',inplace=True)
```

```python
df.drop('Status',axis='columns',inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['BasePay'].fillna(df['BasePay'].mean(),inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(),inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(),inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

Why do I use 'drop' for two features [state, notes]? Because all values for these two features are null, and I don't have any information I can use to fill in the missing values.

Why do I use 'fillna' for these features [basepay, overtimepay, otherpay, benefits]? These features are categorical, and they have a distribution. I believe the best approach for this is to fill in the missing values with the mean.

```python
df['TotalPay']
plt.hist(df['TotalPay'],bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of salaries ')
```

```python
df['JobTitle'].unique()
```

```python
employee_count_by_department = df['JobTitle'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(employee_count_by_department)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
groub=df.groupby('JobTitle').agg({
    'TotalPay': ['count', 'mean', 'min', 'max', 'std']
}).reset_index()
average_salaries = df.groupby('JobTitle')['TotalPay'].mean()
average_salaries.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('Average Salaries Across Departments')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.show()
```
